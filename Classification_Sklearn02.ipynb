## LogisticRegression 을 이용한 분류
- 데이터셋을 학습용, 검증용으로 분리
- LogisticRegression 모델 생성
- from sklearn.linear_model import LogisticRegression
- model = LogisticRegression(max_iter=5000), default 100
- 생성된 모델 학습( logisticRegr.fit(문제, 정답) )
- 검증 : score, predict
- 분류 결과를 시각화 확인

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
model = LogisticRegression(max_iter=5000)
model
```

```python
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(X_train, y_train), model.score(X_test, y_test)
```

```python
pred = model.predict(X_test)
```

### ML 평가지표
*   <img src='https://cwisky.github.io/public/images/precision.jpg' width='40%' align='right'>    
- Accuracy(정확도) : 올바르게 예측된 데이터의 수를 전체 데이터의 수로 나눈 값  
- 태풍이 불 것으로 예보하는 시스템이 항상 태풍이 없다고 판단하면 정확도가 매우 높게 나타남  
- Precision(정밀도) : True으로 예측한 갯수 중에서 실제로 True인 경우의 수  
- Recall(재현율) : 실제 True인 데이터 중에서 모델이 True라고 인식한 수  
- F1 Score(조화평균) : precision, recall을 조합하여 하나의 통계치를 생성한 것    
    
- Accuracy : (TP + TN) / (TP + TN + FP + FN)  
- Precision : TP / (TP + FP)
- Recall : TP / (TP + FN)
- F1 Score : (2 * precision * recall) / (precision - recall)
            

## 포도주 등급 판정 시스템
- LogistiRegression

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df_wine = pd.read_csv('https://cwisky.github.io/public/winequality-red.csv', sep=';')
df_wine
```

```python
df_wine.value_counts('quality').sort_index()
```

```python
df_wine['target'] = df_wine['quality']-3
```

```python
df_wine['target'].value_counts().sort_index()
```

```python
df_wine.drop('quality', axis=1, inplace=True)
df_wine
```

```python
df_wine.isnull().sum()
```

```python
corr_mat = df_wine.corr()
corr_mat
```

```python
import seaborn as sns
plt.figure(figsize=(15, 5))
sns.heatmap(corr_mat, annot=True, linewidth=2)
plt.show()
```

```python
y = df_wine['target'].values
y.shape
```

```python
X = df_wine.drop('target', axis=1).values
X.shape
```

```python
# LogisticRegression을 사용한 다중분류 시스템 작성
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
model = LogisticRegression(max_iter=10000)
model
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train), model.score(X_test, y_test)
```

```python
pred = model.predict(X_test)
```

```python
# alcohol, volatile acidity, quality를 차트에 표시해본다
fig, ax = plt.subplots()
scatter = ax.scatter(df_wine['volatile acidity'], df_wine['alcohol'], c=df_wine['target'], s=2)
legend1 = ax.legend(*scatter.legend_elements(),loc="upper right", title="Classes")
ax.add_artist(legend1)

plt.xlabel('volatile acidity')
plt.ylabel('alcohol')
plt.show()
```

```python
# 학습된 모델이 추정한 등급과 alcohol, volatile acidity를 시각화해본다
fig, ax = plt.subplots()
scatter = ax.scatter(X_test[:,1], X_test[:,10], c=pred, s=2)
legend1 = ax.legend(*scatter.legend_elements(),loc="upper right", title="Classes")
ax.add_artist(legend1)

plt.xlabel('volatile acidity')
plt.ylabel('alcohol')
plt.show()
```
